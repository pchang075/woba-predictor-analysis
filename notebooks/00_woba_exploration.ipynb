```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/raw/2025_batting_stats.csv')
print(df.shape)

df = df[df['PA'] >= 200]
```

```python
plt.hist(df['wOBA'], bins=30)
plt.xlabel('wOBA')
plt.ylabel('Players')
plt.title('Distribution of Weighted On-Base Average for PA ≥ 200')
plt.show()
```

```python
plt.boxplot(df['wOBA'], orientation='horizontal')
plt.xlabel('wOBA')
plt.title('Spread of Weighted On-Base Average for PA ≥ 200')
plt.show()
```

```python
plt.scatter(df['PA'], df['wOBA'], alpha=0.5)
plt.xlabel('PA')
plt.ylabel('wOBA')
plt.title('Weighted On-Base Average vs Plate Appearances for PA ≥ 200')
plt.show()
```
