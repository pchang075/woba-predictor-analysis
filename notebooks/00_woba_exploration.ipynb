```python
import pandas as pd
import matplotlib.pyplot as plt

from pybaseball import batting_stats

df = batting_stats(2025, qual=0)
print(df.shape)
# df.to_csv('../data/batting_stats_2025.csv')

df = df[df['PA']>=200]
```

```python
plt.hist(df['wOBA'], bins=30)
plt.xlabel('wOBA')
plt.ylabel('Number of Players')
plt.title('wOBA Distribution (PA ≥ 200)')
plt.show()
```

```python
plt.boxplot(df['wOBA'], orientation='horizontal')
plt.xlabel('wOBA')
plt.title('wOBA Spread (PA ≥ 200)')
plt.show()
```

```python
plt.scatter(df['PA'], df['wOBA'], alpha=0.5)
plt.xlabel('Plate Appearances')
plt.ylabel('wOBA')
plt.title('wOBA vs Plate Appearances for PA ≥ 200')
plt.show()
```
